4. Llama al eje X "eje X", al eje Y "eje Y" y pon de títulos "recta sube" y "recta baja". Muestra dos etiquetas de ejeX pero solo una de eje Y.

```python
# Creamos los valores del eje X
x = np.linspace(0, 10, 200)

# Rectas
y1 = 3 * x          # recta sube
y2 = -3 * x - 5     # recta baja

# Figura
fig = plt.figure(figsize=(10, 4))

# Dos ejes (izquierda y derecha)
ax1 = fig.add_axes([1, 3, 2, 5])   
ax2 = fig.add_axes([6, 0.5, 0.6, 2])  


ax1.plot(x, y1, color='green')
ax1.set_title('recta sube')
ax1.set_xlabel('eje X')
ax1.set_ylabel('eje Y')    
ax1.grid(True)
ax1.set_xlim(0, 5)
ax1.set_ylim(-15, 15)


ax2.plot(x, y2, color='yellow')
ax2.set_title('recta baja')
ax2.set_xlabel('eje X')
ax2.grid(True)
ax2.set_xlim(0, 5)
ax2.set_ylim(-15, 15)
```

7. Crea un gráfico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sofía mide 190 cm, Carmen mide 170 cm. Las barras serán verdes.

11. Prueba a invertir los ejes con ax.invert_axis()
Necesitarás tener los ejes disponibles fuera de la figura

```python
personas = ["Ana", "Luis", "Pedro", "Sofía", "Carmen"]
alturas = [160, 180, 175, 190, 170]

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(5, 3))

# Gráfico de barras horizontal
ax.barh(personas, alturas, color="blue")


ax.set_title("Altura de un grupo de amigos")
ax.set_xlabel("Altura (cm)")
ax.set_ylabel("Personas")

# Invertir el eje Y
ax.invert_yaxis()   


plt.show()
```

14. Ahora pintamos un gráfico de línea con las ventas respecto al año. Pon un título y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La línea debe ser con rayas y verde.

```python
# Estilo seaborn-white

plt.style.use('seaborn-v0_8-white')


data = {"Year": [2015, 2016, 2017, 2018, 2019],
        "Sold_items_A": [1000, 3500, 4000, 5500, 7000]}

df = pd.DataFrame(data)

# Gráfico de línea (verde y discontinua)

plt.plot(df["Year"], df["Sold_items_A"], color="green", linestyle="--")


plt.title("Ventas de producto A por año")
plt.xlabel("Año")
plt.ylabel("Unidades vendidas")


plt.show()
```
