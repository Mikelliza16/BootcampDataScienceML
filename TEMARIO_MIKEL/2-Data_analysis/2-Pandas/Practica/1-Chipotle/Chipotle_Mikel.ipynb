### Step 4. How many products cost more than $10.00?

```python
import pandas as pd

# Cargar el dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Limpiar la columna de precios: quitar '$' y espacios
chipo["item_price"] = chipo["item_price"].str.replace('$', '', regex=False).str.strip()

# Convertir a float
chipo["item_price"] = chipo["item_price"].astype(float)
```

```python
productos_caros = chipo[chipo['item_price'] > 10.00]
```

```python
num_productos_caros = productos_caros['item_name'].nunique()
print("Número de productos que cuestan más de $10:", num_productos_caros)
```

### Step 4.1: Y cuántos pedidos se han hecho con un producto de más de 10$? Es lo mismo?

```python
# Filtrar productos con precio > $10
productos_caros = chipo[chipo['item_price'] > 10.00]

# Contar cuántos pedidos (order_id) contienen al menos un producto caro
num_pedidos_con_caros = productos_caros['order_id'].nunique()

print("Número de pedidos con al menos un producto > $10:", num_pedidos_con_caros)
```

### Step 4.2: Y cuántos pedidos se han hecho de más de 10$? Es lo mismo?

```python
# Primero, asegurarse de que item_price es float y limpiar símbolos
chipo["item_price"] = chipo["item_price"].str.replace('$', '', regex=False).str.strip().astype(float)

# Calcular el total por pedido (sumando todos los productos de cada order_id)
pedido_total = chipo.groupby('order_id')['item_price'].sum()

# Filtrar pedidos cuyo total > 10
pedidos_mayor_10 = pedido_total[pedido_total > 10]

# Contar cuántos pedidos cumplen esa condición
num_pedidos_mayor_10 = pedidos_mayor_10.count()

print("Número de pedidos con total > $10:", num_pedidos_mayor_10)
```

### Step 4.3: Y en cuántos pedidos se ha pagado más de 10$ por un mismo producto? Es lo mismo?

```python
import pandas as pd

# Asegurarse que item_price es float
chipo["item_price"] = chipo["item_price"].str.replace('$', '', regex=False).str.strip().astype(float)

# Calcular el total por línea (cantidad * precio unitario)
chipo["line_total"] = chipo["quantity"] * chipo["item_price"]

# Filtrar las líneas donde se pagó más de $10 por un mismo producto
lineas_mayor_10 = chipo[chipo["line_total"] > 10]

# Contar cuántos pedidos únicos tienen al menos una línea > $10
num_pedidos_linea_mayor_10 = lineas_mayor_10["order_id"].nunique()

print("Número de pedidos donde se pagó más de $10 por un mismo producto:", num_pedidos_linea_mayor_10)
```
