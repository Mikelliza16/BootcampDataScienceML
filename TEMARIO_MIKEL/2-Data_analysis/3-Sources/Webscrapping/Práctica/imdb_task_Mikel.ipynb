## Web scrapping de IMDB

Descarga la información correspondiente y guarda en un csv el top de las 250 películas mediante webscrapping. Encapsúlalo en un script.

Obtén:
* Título
* Año
* Duración
* Posición
* Rating

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
import time
import pandas as pd

# Abrir la página de IMDb
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/chart/top/")
time.sleep(3)  

# Buscar todas las películas
peliculas = driver.find_elements(By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item")

# Crear archivo CSV
archivo = open("imdb_top250.csv", "w", newline="", encoding="utf-8")
escritor = csv.writer(archivo)
escritor.writerow(["Posición", "Título", "Año", "Duración", "Rating"])

# Recorrer las películas y guardar datos
for i, p in enumerate(peliculas, start=1):
    try:
        titulo = p.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text.split('.', 1)[-1]
    except:
        titulo = ""
    try:
        año = p.find_element(By.CSS_SELECTOR, "span.sc-b0691f29-8").text
    except:
        año = ""
    try:
        duracion = p.find_element(By.CSS_SELECTOR, "span.sc-b0691f29-7").text
    except:
        duracion = ""
    try:
        rating = p.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--imdb").text.replace("★", "")
    except:
        rating = ""
    
    escritor.writerow([i, titulo, año, duracion, rating])

# Cerrar todo
archivo.close()
driver.quit()
```

```python
pf = pd.read_csv("imdb_top250.csv")
```
