```python
import pandas as pd
import io

ruta = "version2_avalanche_accidents_all_switzerland_since_1970.csv"

# leer todo y saltar las tres líneas de cabecera informativa
with open(ruta, "r", encoding="utf-8") as f:
    lines = f.readlines()[3:]

# --- limpiar la cabecera real (línea 0 del bloque leído) ---
header_line = lines[0].strip()

# quita comillas inicial/final y dupes "" -> "
if header_line.startswith('"') and header_line.endswith('"'):
    header_line = header_line[1:-1]
header_line = header_line.replace('""', '"')

# sustituir la cabecera en las líneas
lines[0] = header_line + "\n"

# --- ahora pandas puede leerlo sin errores ---
dfavalanchas = pd.read_csv(io.StringIO("".join(lines)), sep=",", engine="python")

# convertir tipos básicos
dfavalanchas["date"] = pd.to_datetime(dfavalanchas["date"], errors="coerce")
for c in ["start.zone.coordinates.latitude",
          "start.zone.coordinates.longitude",
          "start.zone.elevation",
          "number.dead", "number.caught", "number.fully.buried"]:
    if c in dfavalanchas.columns:
        dfavalanchas[c] = pd.to_numeric(dfavalanchas[c], errors="coerce")

dfavalanchas.head()

```

```python
import pycountry

# --- Diccionario de excepciones manuales ---
manual_fixes = {
    "United States": "United States of America",
    "Russia": "Russian Federation",
    "South Korea": "Korea, Republic of",
    "North Korea": "Korea, Democratic People's Republic of",
    "Vietnam": "Viet Nam",
    "Congo": "Congo, Republic of",
    "Democratic Republic of the Congo": "Congo, Democratic Republic of",
    "Iran": "Iran, Islamic Republic of",
    "Syria": "Syrian Arab Republic",
    "Venezuela": "Venezuela, Bolivarian Republic of",
    "Tanzania": "Tanzania, United Republic of",
}

# --- Función mejorada para obtener ISO3 ---
def get_iso3(name):
    # Aplicar corrección manual si existe
    if name in manual_fixes:
        name = manual_fixes[name]
    try:
        return pycountry.countries.lookup(name).alpha_3
    except:
        return None

# --- Recalcular ISO3 ---
volc_country["ISO3"] = volc_country["Country"].apply(get_iso3)
pop_2020["ISO3"] = pop_2020["Country Code"]

# --- Nuevo merge ---
df = pd.merge(volc_country, pop_2020, on="ISO3", how="inner")

# --- Limpiar columnas ---
df = df[["Country_x", "Volcano_Count", "Population_2020", "ISO3"]].rename(columns={"Country_x": "Country"})

print(f"✅ Países combinados correctamente: {df.shape[0]}")
print("✅ Países añadidos con corrección manual:", list(manual_fixes.keys()))
df[df["Country"].isin(list(manual_fixes.keys()))]
```
