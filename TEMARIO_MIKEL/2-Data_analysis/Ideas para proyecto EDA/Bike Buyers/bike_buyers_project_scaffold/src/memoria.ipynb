```
from src.utils.data_loading import load_bike_buyers
import pandas as pd

# üëâ Ajusta el nombre del archivo CSV
CSV_PATH = 'src/data/Bike_Buyers.csv'
df = load_bike_buyers(CSV_PATH)
df.head()
```

## 5. An√°lisis exploratorio y validaci√≥n de hip√≥tesis

### H1 ‚Äî G√©nero vs Compra

```
from src.utils.stats import chi2_table
from src.utils.plotting import countplot_ratio

# Aseg√∫rate de que 'purchased_bike' y 'gender' existan con estos nombres
res = chi2_table(df, 'gender', 'purchased_bike')
print(res['table'])
print('chi2=', res['chi2'], 'p=', res['p'])
countplot_ratio(df, 'gender', 'purchased_bike')
```

### H2 ‚Äî Edad vs Compra

```
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(data=df, x='purchased_bike', y='age')
plt.title('Edad por compra de bicicleta')
plt.show()
```

### H3 ‚Äî Ingreso vs Compra

```
sns.boxplot(data=df, x='purchased_bike', y='income')
plt.title('Ingreso por compra de bicicleta')
plt.show()
```

### H4 ‚Äî Hijos vs Compra

```
countplot_ratio(df, 'children', 'purchased_bike')
```

### H5 ‚Äî N√∫mero de coches vs Compra

```
countplot_ratio(df, 'cars', 'purchased_bike')
```

### H6 ‚Äî Distancia al trabajo vs Compra

```
countplot_ratio(df, 'commute_distance', 'purchased_bike')
```
