```python
import pandas as pd
import glob
import os

# 1. Definir la ruta donde están los CSVs descomprimidos
path = 'data/raw/' 
all_files = glob.glob(os.path.join(path, "*.csv"))

# Lista para guardar cada dataframe temporal
lista_dfs = []

# 2. Bucle para leer cada archivo y añadir la columna "Brand" (Marca)
for filename in all_files:
    # Leemos el CSV
    df_temp = pd.read_csv(filename)
    
    # Extraemos el nombre del archivo (ej: 'audi.csv' -> 'audi') para usarlo como marca
    marca = os.path.basename(filename).split('.')[0].capitalize()
    
    # Creamos la columna nueva (Requisito clave para tener >10 columnas)
    df_temp['brand'] = marca
    
    # Añadimos a la lista
    lista_dfs.append(df_temp)

# 3. Unir todo en un solo DataFrame gigante
df_final = pd.concat(lista_dfs, axis=0, ignore_index=True)

# 4. Feature Engineering rápida (para asegurar el tiro)
# Creamos la antigüedad
df_final['car_age'] = 2024 - df_final['year']

# Convertimos millas a km (opcional, pero queda profesional)
df_final['kms_driven'] = df_final['mileage'] * 1.60934

# 5. Guardar en processed
df_final.to_csv('data/processed/dataset_coches_completo.csv', index=False)

# Verificación
print(f"Dimensiones del dataset: {df_final.shape}")
print(f"Columnas finales: {df_final.columns.tolist()}")
```
